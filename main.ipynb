```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure and axis
fig, ax = plt.subplots(figsize=(12, 8))

# Data points based on the chart
# [Beta to NIFTY, Net Total Return, size, color, label]
data_points = [
    [0.25, 7.2, 300, '#2c5f7a', '1'],      # Point 1
    [0.65, 10.5, 300, '#2c5f7a', '2'],     # Point 2  
    [0.35, 10.0, 300, '#2c5f7a', '3'],     # Point 3
    [0.22, 12.0, 300, '#2c5f7a', '4'],     # Point 4
    [0.18, 14.5, 200, '#cccccc', 'Reference: typical US fund'],  # Reference point
    [0.95, 14.5, 300, '#2d5f2d', 'PMS Long-Only Equity']  # PMS point
]

# Plot each point
for i, (beta, returns, size, color, label) in enumerate(data_points):
    if i < 4:  # Numbered points
        ax.scatter(beta, returns, s=size, c=color, alpha=0.8, zorder=3)
        ax.text(beta, returns, label, ha='center', va='center', 
                color='white', fontsize=12, fontweight='bold', zorder=4)
    else:  # Reference and PMS points
        ax.scatter(beta, returns, s=size, c=color, alpha=0.8, zorder=3)
        if 'Reference' in label:
            ax.annotate(label, (beta, returns), xytext=(beta-0.05, returns+0.8),
                       fontsize=10, color='gray', ha='center')
        else:
            ax.annotate(label, (beta, returns), xytext=(beta-0.05, returns+0.8),
                       fontsize=10, color='green', ha='center', fontweight='bold')

# Draw arrows between numbered points to show progression
arrows = [
    (0.25, 7.2, 0.65, 10.5),   # 1 to 2
    (0.65, 10.5, 0.35, 10.0),  # 2 to 3
    (0.35, 10.0, 0.22, 12.0)   # 3 to 4
]

for x1, y1, x2, y2 in arrows:
    ax.annotate('', xy=(x2, y2), xytext=(x1, y1),
                arrowprops=dict(arrowstyle='->', color='#2c5f7a', lw=2, alpha=0.7))

# Customize the plot
ax.set_xlabel('Beta to NIFTY', fontsize=14, fontweight='bold', color='gray')
ax.set_ylabel('Net Total Return', fontsize=14, fontweight='bold', color='gray')
ax.set_title('Amunra Advisors LLP', fontsize=18, fontweight='bold', pad=20)

# Set axis limits and formatting
ax.set_xlim(0, 1.05)
ax.set_ylim(0, 16)

# Format y-axis as percentages
y_ticks = np.arange(0, 17, 2)
ax.set_yticks(y_ticks)
ax.set_yticklabels([f'{int(y)}%' for y in y_ticks])

# Format x-axis
x_ticks = np.arange(0, 1.1, 0.20)
ax.set_xticks(x_ticks)
ax.set_xticklabels([f'{x:.2f}' for x in x_ticks])

# Grid styling
ax.grid(True, alpha=0.3, linestyle='-', linewidth=0.5)
ax.set_facecolor('white')

# Remove top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('gray')
ax.spines['bottom'].set_color('gray')

# Adjust layout and display
plt.tight_layout()
plt.show()

# Optional: Save the figure
# plt.savefig('amunra_advisors_chart.png', dpi=300, bbox_inches='tight')
```
