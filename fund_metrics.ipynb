##### Deriv Strategy - simulated return

```python
def simulate_portfolio(df, target_return_ann=0.14, beta=0.2):
    """
    df : pd.DataFrame with columns ["Market", "Cash"] containing monthly returns (as decimals, e.g. 0.01 = 1%)
    target_return_ann : target annualized return (e.g. 0.14 for 14%)
    beta : desired beta to market
    """
    n_months = len(df)
    periods_per_year = 12
    
    # Convert annual to monthly return
    target_return_monthly = (1 + target_return_ann)**(1/periods_per_year) - 1
    
    # Expected monthly values from series
    market_mean = df["market_return"].mean()
    cash_mean = df["cash_return"].mean()
    
    # Solve for alpha
    alpha = target_return_monthly - (beta * market_mean + (1 - beta) * cash_mean)
    
    # Construct portfolio returns series
    df["gross_return"] = alpha + beta * df["market_return"] + (1 - beta) * df["cash_return"]
    
    return df, alpha
```

```python
file_path = os.getcwd() + '/data/prop_strat.xlsx'
df = pd.read_excel(file_path)
fund_data, alpha = simulate_portfolio(df, target_return_ann=0.14, beta=0.2) # Create the simulated fund data with 20% beta
fund_data.set_index('date', inplace=True)
```

```python
# Initialize analytics
portfolio = PortfolioAnalytics(fund_data)

# Generate comprehensive report
results = portfolio.generate_report()

# Create plots
fig = portfolio.create_plots()
plt.show()

# Export to Excel
portfolio.export_results('portfolio_analysis.xlsx')
```
